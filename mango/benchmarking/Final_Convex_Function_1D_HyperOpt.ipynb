# Example of optimizing a convex function

# Goal is to test the objective values found by HyperOpt

```python
from hyperopt import fmin, tpe, hp, STATUS_OK, Trials
```

```python
Store_Optimal_X = []
Store_Results = []
num_of_tries = 100
```

```python
def objective(x):
    return {'loss': x ** 2, 'status': STATUS_OK }
```

```python

#Since 5 random were tried in Bayesian, here we add 5 extra

for i in range(num_of_tries):
    best_20 = fmin(fn=objective, space=hp.uniform('x', -5000, 5000), algo=tpe.suggest, max_evals=25)
    #best_30 = fmin(fn=objective, space=hp.uniform('x', -5000, 5000), algo=tpe.suggest, max_evals=35)
    best_40 = fmin(fn=objective, space=hp.uniform('x', -5000, 5000), algo=tpe.suggest, max_evals=45)
    #best_50 = fmin(fn=objective, space=hp.uniform('x', -5000, 5000), algo=tpe.suggest, max_evals=55)
    best_60 = fmin(fn=objective, space=hp.uniform('x', -5000, 5000), algo=tpe.suggest, max_evals=65)
    #best_70 = fmin(fn=objective, space=hp.uniform('x', -5000, 5000), algo=tpe.suggest, max_evals=75)
    best_80 = fmin(fn=objective, space=hp.uniform('x', -5000, 5000), algo=tpe.suggest, max_evals=85)
    #best_90 = fmin(fn=objective, space=hp.uniform('x', -5000, 5000), algo=tpe.suggest, max_evals=95)
    best_100 = fmin(fn=objective, space=hp.uniform('x', -5000, 5000), algo=tpe.suggest, max_evals=105)
    best_120 = fmin(fn=objective, space=hp.uniform('x', -5000, 5000), algo=tpe.suggest, max_evals=125)
    best_140 = fmin(fn=objective, space=hp.uniform('x', -5000, 5000), algo=tpe.suggest, max_evals=145)
    best_160 = fmin(fn=objective, space=hp.uniform('x', -5000, 5000), algo=tpe.suggest, max_evals=165)
    best_180 = fmin(fn=objective, space=hp.uniform('x', -5000, 5000), algo=tpe.suggest, max_evals=185)
    best_200 = fmin(fn=objective, space=hp.uniform('x', -5000, 5000), algo=tpe.suggest, max_evals=205)
    print(i,':',[best_20['x'],best_40['x'],best_60['x']])
    Store_Optimal_X.append([best_20['x'],best_40['x'],best_60['x'],best_80['x'],best_100['x'],best_120['x'],best_140['x'],best_160['x'],best_180['x'],best_200['x']])
```

```python
import numpy as np
Store_Optimal_X=np.array(Store_Optimal_X)
```

```python
Store_Optimal_X.shape
```

```python
import numpy as np

plotting_itr =[20,40,60,80,100,120,140,160,180,200]
```

```python
Y = []
#count range between -1 and 1 and show it
for i in range(len(plotting_itr)):
    x_value = Store_Optimal_X[:,i]
    #print(len(x_value))
    y_value = -(x_value*x_value)
    y_value = np.count_nonzero((y_value > -100))/y_value.size
    print(y_value)
    #print(len(x_value))
    #y_value = ()
    Y.append(y_value*100)
```

```python
import numpy as np
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(10,10))


plt.plot(plotting_itr,Y,label = 'HyperOpt',linewidth=3.0) #x, y


plt.xlabel('Number of Iterations',fontsize=25)
plt.ylabel(' % of $Fun_{value}$ in (-100,0)',fontsize=25)
plt.title('Variation of Optimal Value of X with iterations',fontsize=20)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.yticks(np.arange(10, 110, step=10))
plt.xticks(np.arange(20, 220, step=20))
plt.grid(True)
plt.legend(fontsize=20)
plt.show()
```
