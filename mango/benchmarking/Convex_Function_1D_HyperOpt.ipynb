# Example of optimizing a convex function

# Goal is to test the objective values found by HyperOpt

```python
from hyperopt import fmin, tpe, hp, STATUS_OK, Trials
```

```python
Store_Optimal_X = []
Store_Results = []
num_of_tries = 100
```

```python
def objective(x):
    return {'loss': x ** 2, 'status': STATUS_OK }
```

```python

#Since 5 random were tried in Bayesian, here we add 5 extra

for i in range(num_of_tries):
    best_20 = fmin(fn=objective, space=hp.uniform('x', -5000, 5000), algo=tpe.suggest, max_evals=25)
    best_30 = fmin(fn=objective, space=hp.uniform('x', -5000, 5000), algo=tpe.suggest, max_evals=35)
    best_40 = fmin(fn=objective, space=hp.uniform('x', -5000, 5000), algo=tpe.suggest, max_evals=45)
    best_60 = fmin(fn=objective, space=hp.uniform('x', -5000, 5000), algo=tpe.suggest, max_evals=65)
    print(i,':',[best_20['x'],best_30['x'],best_40['x'],best_60['x']])
    Store_Optimal_X.append([best_20['x'],best_30['x'],best_40['x'],best_60['x']])
```

```python
import numpy as np
Store_Optimal_X=np.array(Store_Optimal_X)
```

```python
import numpy as np
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(10,10))
n, bins, patches = plt.hist(Store_Optimal_X[:,0], 30, facecolor='g', alpha=0.75)
def autolabel(rects):
    """
    Attach a text label above each bar displaying its height
    """
    for rect in rects:
        height = rect.get_height()
        plt.text(rect.get_x() + rect.get_width()/2., 1.0*height,
                '%d' % int(height),
                ha='center', va='bottom',fontsize=15)
plt.xlabel('X-Value',fontsize=25)
plt.ylabel('Number of Occurence',fontsize=25)
plt.title('Hyperopt Optimal Objective: Iterations 20',fontsize=20)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.grid(True)
autolabel(patches)
plt.show()
```
